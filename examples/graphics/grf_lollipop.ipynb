```r
#color palette
colors <- brewer.pal(4, 'Set1')

# This function is used only to set graphics size in this notebook. 
plot.size(10, 5)

# setting the font size for all charts
font <- theme(text = element_text(size=16))
```

```r
#iris dataset for the example
head(iris)
```

```r
data <- iris |> group_by(Species) |> summarize(Sepal.Length=mean(Sepal.Length))
head(data)
```

## Lollipop plot

The lollipop graph has the same goal as a bar graph.

```r
grf <- plot.lollipop(data, colors=colors[1], max_value_gap=0.2, flip=FALSE) + font
plot(grf)
```

```r
grf <- plot.lollipop(data, colors=colors[1], max_value_gap=0.1) + font
plot(grf)
```

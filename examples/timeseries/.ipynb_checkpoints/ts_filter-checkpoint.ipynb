## Time series preprocessing

```r
# DAL Library
# version 2.1

source("../dal_transform.R")
source("../ts_data.R")
source("../ts_filter.R")
```

### Cosine series for studying

```r
i <- seq(0, 2*pi+8*pi/50, pi/50)
x <- cos(i)

options(repr.plot.width=6, repr.plot.height=5)  
par(mfrow = c(1, 1))
plot(i, x)
lines(i, x)
```

### sliding windows

```r
sw_size <- 10
xw <- ts_data(x, sw_size)
i <- 1:nrow(xw)
y <- xw[,sw_size]

plot(i, y)
lines(i, y)
```

### smooth

```r
filter <- ts_smooth()
xa <- transform(filter, xw)
idx <- attr(xa, "idx")

plot(x = i, y = y, main = "cosine")
lines(x = i, y = y, col="black")
lines(x = idx, y = xa[,sw_size], col="green")
```

### moving average

```r
filter <- ts_ma(3)
xa <- transform(filter, xw)
idx <- attr(xa, "idx")

plot(x = i, y = y, main = "cosine")
lines(x = i, y = y, col="black")
for (j in 1:nrow(xa)) {
  lines(x = (idx[j]-sw_size+1):idx[j], y = xa[j,1:sw_size], col="green")
}
```

### exponential moving average

```r
filter <- ts_ema(3)
xa <- transform(filter, xw)
idx <- attr(xa, "idx")

plot(x = i, y = y, main = "cosine")
lines(x = i, y = y, col="black")
for (j in 1:nrow(xa)) {
  lines(x = (idx[j]-sw_size+1):idx[j], y = xa[j,1:sw_size], col="green")
}
```

### awareness

```r
filter <- ts_awareness(0.25)
xa <- transform(filter, xw)
idx <- attr(xa, "idx")

plot(x = i, y = y, main = "cosine")
lines(x = i, y = y, col="black")
for (j in 1:nrow(xa)) {
lines(x = (idx[j]-sw_size+1):idx[j], y = xa[j,1:sw_size], col="green")
}
```

### smooth awareness

```r
filter <- ts_aware_smooth(0.25)
xa <- transform(filter, xw)
idx <- attr(xa, "idx")

plot(x = i, y = y, main = "cosine")
lines(x = i, y = y, col="black")
for (j in 1:nrow(xa)) {
lines(x = (idx[j]-sw_size+1):idx[j], y = xa[j,1:sw_size], col="green")
}
```
